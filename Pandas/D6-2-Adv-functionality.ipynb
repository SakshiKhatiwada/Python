```python
import pandas as pd
```

```python
import numpy as np
```

```python
coffee = pd.read_csv("complete-pandas-tutorial/warmup-data/coffee.csv")
```

```python
coffee['price'] = np.where(coffee["Coffee Type"] == "Latte", 45, 35)
coffee['revenue'] = coffee['Units Sold'] * coffee['price']
```

```python
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
```

```python
coffee['pct_change'] = coffee['revenue'] / coffee['yesterday_revenue'] * 100
```

```python
bios["height_rank"] = bios["height_rank"].rank(ascending=False)
```

```python
coffee['cum_rev'] = coffee['revenue'].cumsum()
```

```python
latte = coffee[coffee['Coffee Type'] == "Latte"].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum() # NOTE: rolling is cummulative freq of 3 rows, rolling() is a method of the Series object, not the DataFrame object
```

#NOTE - : A Series is essentially a single column within a DataFrame, so they share many of the same methods and attributes. However, a Series has only one column index (its own index), whereas a DataFrame has both row indices and column headers.
