```python
import pandas as pd
```

```python
bios = pd.read_csv("complete-pandas-tutorial/data/bios.csv")
```

```python
nocs = pd.read_csv("complete-pandas-tutorial/data/noc_regions.csv")
```

```python
nocs.head()
```

```python
bios.head()
```

```python
# bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left') # INFO it is giving NOC_x and NOC_y because of same name, we can specify suffixes as:
bios_new = pd.merge(
    bios,
    nocs,
    left_on="born_country",
    right_on="NOC",
    how="left",
    suffixes=["bios", "noc"],
)
#NOTE because the columns had different names, else directly on="" is used
```

```python
bios_new.rename(columns={"region" :"born_country_full"}, inplace=True) #INFO to retain the value even on next lines
```

```python
bios_new[bios_new["NOCbios"] != bios_new["born_country_full"]]
# bios_new
```

```python
usa = bios[bios["born_country"] == "USA"].copy()
gbr = bios[bios["born_country"] == "GBR"].copy()
```

```python
new_df = pd.concat([usa,gbr])
```
