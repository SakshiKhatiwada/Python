# Updating Data Vaues

```python
#SECTION updating data values

coffee.loc[1:3, "Units Sold"] = 100

coffee.at[0,"Units Sold"] = 37
coffee.iat[0,2] = 47
coffee
```

```python
# SECTION - accessing data

# coffee.Day # for single word
# coffee["Day"] # for single value
coffee[["Day", "Coffee Type"]]  # for multiple values, lists
```

```python
#SECTION - sorting

coffee.sort_values("Units Sold", ascending=False)
coffee.sort_values(["Units Sold", "Coffee Type"], ascending=False) #multiple constraint
```
