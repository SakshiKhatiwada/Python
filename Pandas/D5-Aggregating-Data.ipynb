```python
import numpy as np
```

```python
bios = pd.read_csv("complete-pandas-tutorial/data/bios.csv")
```

```python
coffee = pd.read_csv("complete-pandas-tutorial/warmup-data/coffee.csv")
```

```python
# bios['born_city'].value_counts()
bios[bios['born_country'] == "USA"]['born_region'].value_counts().tail()
```

```python
# coffee.drop(columns={"price"}, inplace=True)
coffee['price'] = np.where(coffee['Coffee Type'] == "Latte", 35, 45)
```

```python
coffee
```

```python
# coffee.groupby(['Coffee Type'])['Units Sold'].mean()
# coffee.groupby(['Coffee Type']).agg({'Units Sold':'sum', 'price': 'mean'}) #aggregate, find sum for Units Sold, and mean for price
coffee.groupby(['Coffee Type','Day']).agg({'Units Sold':'sum', 'price': 'mean'}) #aggregate, find sum for Units Sold, and mean for price
```

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['price']
```

```python
coffee
```

```python
pivot = coffee.pivot(columns='Coffee Type', index='Day', values="revenue")
```

```python
# pivot
pivot.sum()
# pivot.loc['Monday', 'Latte']
pivot.sum(axis=1)
```
