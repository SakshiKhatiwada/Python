```python
import yfinance as yf
```

```python
import numpy as np
import pandas as pd, matplotlib.pyplot as plt
```

```python
apple = yf.Ticker("AAPL")
```

```python
data = apple.history(period='3mo')
data.head()
```

#### To visualize the raw data

```python
plt.plot(data["Close"])

plt.title(
    "AAPL Closing Price",
    fontdict={
        "fontsize": "20",
        "fontname": "Comic Sans MS",
        "fontweight": "light",
        "color": "black",
    },
)

plt.xlabel("Data")
plt.xticks(rotation=45)

plt.ylabel("Price")


plt.show()
```

#### Calculate Moving Averages for Time-Series Analysis 

```python
data['MA20'] = data["Close"].rolling(window=20).mean()
data['MA50'] = data["Close"].rolling(window=50).mean()

#INFO sthis rolling().mean() are from Pandas
```

```python
plt.figure(figsize=(8, 5))
plt.style.use("ggplot")
# plots
plt.plot(data["MA20"], "go-", label="Moving Average of 20")
plt.plot(data["MA50"], "r^-", label="Moving Average of 50")
plt.plot(data["Close"], "bv--", label="Closing Price")
plt.plot(data["Volatility"], "y>-", label="volatility")

plt.legend()

# labels
plt.title("Time-Series Analysis")
plt.xlabel("Data")
plt.ylabel("Price")
plt.xticks(rotation=45)

plt.show()
```

NOTE: MA50 will start later on the chart, as it requires more data to get the average over 50 days.

MA20 will be plotted earlier since it uses fewer data points (only 20), and the average can be calculated faster.

#### Volatility

```python
data["Volatility"] = data['Close'].rolling(window=20).std()
```

```python
plt.plot(data['Volatility'])
```
