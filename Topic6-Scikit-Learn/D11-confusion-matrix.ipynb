### How to generate and interpret confusion matrices using Scikit-Learn with the Iris dataset?

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# loading data set

flower_iris = load_iris()
X, y = flower_iris.data, flower_iris.target
```

```python
# split the data, training vs testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=34
)
```

```python
# Train a K-Nearest Neighbors Classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
# make predictions
y_pred = knn.predict(X_test)
```

```python
# generating confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(flower_iris.target_names) # ['setosa' 'versicolor' 'virginica']
print("Confusion Matrix: \n", cm)
```

```python
# to visualize the confusion matrix:
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=flower_iris.target_names)
display.plot()
plt.show()
```
